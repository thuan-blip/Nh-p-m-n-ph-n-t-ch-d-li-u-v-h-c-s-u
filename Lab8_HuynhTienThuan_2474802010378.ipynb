Câu 4

```
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Tải trực tiếp các gói cần thiết ngay trong ô này để tránh lỗi mất gói dữ liệu
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)  # Dự phòng cho các phiên bản NLTK mới nhất
nltk.download('stopwords', quiet=True)

text = "This is a simple example to demonstrate the removal of stopwords in NLTK."

# Tiến hành tách từ
tokens = word_tokenize(text)

# Lấy danh sách stopword tiếng Anh
english_stops = set(stopwords.words('english'))

# Lọc các từ không phải là stopword
filtered_words = [word for word in tokens if word.lower() not in english_stops]

print("Văn bản gốc:", tokens)
print("\nSau khi loại bỏ stopword:", filtered_words)
```

Câu 5

```
from nltk.corpus import stopwords

all_stopwords = set(stopwords.words('english'))

# Giả sử ta muốn giữ lại các từ phủ định (bỏ chúng ra khỏi danh sách từ dừng)
words_to_keep = {'not', 'no', 'nor'}
custom_stopwords = all_stopwords - words_to_keep

print("Số lượng stopword ban đầu:", len(all_stopwords))
print("Số lượng stopword sau khi loại bớt từ phủ định:", len(custom_stopwords))
print("Kiểm tra từ 'not' trong danh sách mới (False nghĩa là 'not' sẽ KHÔNG bị lọc bỏ):", 'not' in custom_stopwords)
```

```
from nltk.corpus import wordnet

word = "computer"
synsets = wordnet.synsets(word)

if synsets:
    print(f"Định nghĩa của '{word}':", synsets[0].definition())
    print(f"Ví dụ của '{word}':", synsets[0].examples())
else:
    print("Không tìm thấy từ này trong WordNet.")
```

Câu 7

```
from nltk.corpus import wordnet

synonyms = set()
antonyms = set()
word = "good"

for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name()) # Thêm từ đồng nghĩa
        if lemma.antonyms():          # Nếu có từ trái nghĩa
            antonyms.add(lemma.antonyms()[0].name())

print(f"Các từ đồng nghĩa với '{word}':", list(synonyms))
print(f"Các từ trái nghĩa với '{word}':", list(antonyms))
```

Câu 9

```
from nltk.corpus import wordnet

# Lấy nghĩa danh từ (n - noun) đầu tiên (.01) của cat và dog
cat = wordnet.synset('cat.n.01')
dog = wordnet.synset('dog.n.01')

similarity = cat.wup_similarity(dog)
print(f"Độ giống nhau ngữ nghĩa giữa 'cat' và 'dog': {similarity:.2f}")
```

Câu 12

```
import random
from nltk.corpus import names

# Tạo danh sách các tuple chứa (Tên, Giới tính)
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Xáo trộn danh sách ngẫu nhiên với seed cố định
random.seed(42)
random.shuffle(labeled_names)

print("15 tên được gắn nhãn ngẫu nhiên:")
for name, gender in labeled_names[:15]:
    print(f"{name}: {gender}")
```

```
import random
from nltk.corpus import names

# Tạo lại danh sách gộp để đảm bảo ô lệnh chạy độc lập không bị lỗi
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Rút trích chữ cái cuối cùng và tạo mảng đặc trưng mới (Ký tự cuối, Giới tính)
featuresets = [(name[-1].lower(), gender) for (name, gender) in labeled_names]

print("10 đặc trưng (Ký tự cuối, Giới tính) đầu tiên trích xuất được:")
for feature in featuresets[:10]:
    print(feature)
```

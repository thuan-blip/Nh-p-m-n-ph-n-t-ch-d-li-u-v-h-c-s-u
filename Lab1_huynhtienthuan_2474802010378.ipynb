Nội dung: Thao tác dữ liệu điểm thi đại học của học sinh được cho bởi bảng bên dưới
Mục tiêu: Sinh viên đạt được kiến thức sau
1. Tìm hiểu nghiệp vụ dữ liệu
2. Nhập liệu bằng công cụ từ file excel
3. Xác định dữ liệu định tính và định lượng
4. Hiệu chỉnh các thang đo phù hợp và kiểu giá trị dữ liệu cho từng biến số
5. Hiệu chỉnh dữ liệu và xử lý dữ liệu thiếu
6. Chuyển đổi (transformation) dữ liệu theo khoảng cho trước
7. Tạo biến số phụ thuộc theo biến độc lập
8. Tạo biến định tính phân loại
Dữ liệu dulieuxettuyendaihoc.csv được mô tả như sau
Dữ liệu lưu trữ điểm trung bình môn, khu vực, khối thi và điểm thi đại học của 100 học sinh.
- T1, L1, H1, S1, V1, X1, D1, N1 lần lượt là điểm trung bình các môn Toán, Lý ,Hóa,
Sinh, Văn, Sử, Địa, Ngoại ngữ năm lớp 10
- T2, L2, H2, S2, V2, X2, D2, N2 lần lượt là điểm trung bình các môn Toán, Lý ,Hóa,
Sinh, Văn, Sử, Địa, Ngoại ngữ năm lớp 11
- T6, L6, H6, S6, V6, X6, D6, N6 lần lượt là điểm trung bình các môn Toán, Lý ,Hóa,
Sinh, Văn, Sử, Địa, Ngoại ngữ năm lớp 12
- GT: Giới tính
- DT: Dân tộc
- KV, KT lần lượt là khu vực thi và khối thi
- DH1, DH2, DH3 lần lượt là điểm thi đại học môn 1, môn 2, môn 3

câu 1 + câu 2

```python
import pandas as pd

df = pd.read_csv("dulieuxettuyendaihoc.csv")

print("Dữ liệu định lượng:")
print(df.select_dtypes(include="number").columns)

print("\nDữ liệu định tính:")
print(df.select_dtypes(exclude="number").columns)
```

câu 3

```python
print("10 dòng đầu:")
print(df.head(10))

print("\n10 dòng cuối:")
print(df.tail(10))
```

câu 4

```python
print("Số dữ liệu thiếu:")
print(df["DT"].isnull().sum())

print("\nGiá trị khác nhau:")
print(df["DT"].unique())

df["DT"] = df["DT"].fillna(0)

print("\nSau khi xử lý:")
print(df["DT"].isnull().sum())
```

câu 5

```python
print("Số dữ liệu thiếu:")
print(df["T1"].isnull().sum())

mean_T1 = df["T1"].mean()

df["T1"] = df["T1"].fillna(mean_T1)

print("\nMean =", mean_T1)

print("\nSau khi xử lý:")
print(df["T1"].isnull().sum())
```

câu 6

```python
cot_diem = [
    "T1","L1","H1","S1","V1","X1","D1","N1",
    "T2","L2","H2","S2","V2","X2","D2","N2",
    "T6","L6","H6","S6","V6","X6","D6","N6",
    "DH1","DH2","DH3"
]

for cot in cot_diem:
    df[cot] = df[cot].fillna(df[cot].mean())

print("Đã xử lý dữ liệu thiếu")
```

câu 7

```python
df["TBM1"] = (
    df["T1"]*2 + df["L1"] + df["H1"] + df["S1"] +
    df["V1"]*2 + df["X1"] + df["D1"] + df["N1"]
) / 10

df["TBM2"] = (
    df["T2"]*2 + df["L2"] + df["H2"] + df["S2"] +
    df["V2"]*2 + df["X2"] + df["D2"] + df["N2"]
) / 10

df["TBM3"] = (
    df["T6"]*2 + df["L6"] + df["H6"] + df["S6"] +
    df["V6"]*2 + df["X6"] + df["D6"] + df["N6"]
) / 10

print(df[["TBM1","TBM2","TBM3"]].head())
```

câu 8

```python
def xep_loai(diem):

    if diem < 5:
        return "Y"

    elif diem < 6.5:
        return "TB"

    elif diem < 8:
        return "K"

    elif diem < 9:
        return "G"

    else:
        return "XS"

df["XL1"] = df["TBM1"].apply(xep_loai)
df["XL2"] = df["TBM2"].apply(xep_loai)
df["XL3"] = df["TBM3"].apply(xep_loai)

print(df[["XL1","XL2","XL3"]].head())
```

câu 9

```python
df["US_TBM1"] = (
    (df["TBM1"] - df["TBM1"].min()) /
    (df["TBM1"].max() - df["TBM1"].min())
) * 4

df["US_TBM2"] = (
    (df["TBM2"] - df["TBM2"].min()) /
    (df["TBM2"].max() - df["TBM2"].min())
) * 4

df["US_TBM3"] = (
    (df["TBM3"] - df["TBM3"].min()) /
    (df["TBM3"].max() - df["TBM3"].min())
) * 4

print(df[["US_TBM1","US_TBM2","US_TBM3"]].head())
```

câu 10

```python
def ket_qua(row):

    if row["KT"] == "A" or row["KT"] == "A1":
        diem = (row["DH1"]*2 + row["DH2"] + row["DH3"]) / 4

    elif row["KT"] == "B":
        diem = (row["DH1"] + row["DH2"]*2 + row["DH3"]) / 4

    else:
        diem = (row["DH1"] + row["DH2"] + row["DH3"]) / 3

    if diem >= 5:
        return 1
    else:
        return 0

df["KQXT"] = df.apply(ket_qua, axis=1)

print(df[["KT","KQXT"]].head())
```
